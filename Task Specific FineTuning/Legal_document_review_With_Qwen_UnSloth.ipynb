```python
import unsloth
from unsloth import FastLanguageModel
import torch
```

```python
max_seq_length = 8192
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Qwen2.5-14B-Instruct-bnb-4bit",
    max_seq_length=max_seq_length,
    dtype=None,
    load_in_4bit=True,
    device_map="auto",
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=64,
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ],
    lora_alpha=64,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=3407,
    use_rslora=False,
)
```

```python
from datasets import load_dataset
dataset = load_dataset("pile-of-law/pile-of-law", "r_legaladvice", split="train[:3000]")
```

```python
def format_legal_analysis(examples):
    """Format examples for legal document analysis using actual dataset structure"""
    texts = []

    for i in range(len(examples["text"])):
        legal_text = examples["text"][i].strip()

        # Skip very short texts that won't be useful for legal training
        if len(legal_text) < 100:
            continue

        # Create legal analysis format from r/legaladvice posts
        # These are typically structured as: Title/Question + Details + Comments/Advice

        # Try to split into question and advice parts
        paragraphs = legal_text.split('\n\n')

        if len(paragraphs) >= 2:
            # Use first paragraph as legal question, rest as analysis
            legal_question = paragraphs[0]
            legal_response = '\n\n'.join(paragraphs[1:])
        else:
            # For shorter texts, create a legal analysis prompt
            legal_question = legal_text[:400] + "..." if len(legal_text) > 400 else legal_text
            legal_response = "This legal matter requires careful analysis considering applicable laws, regulations, and legal precedents. It's important to evaluate all relevant factors and potential legal implications before proceeding."

        # Create comprehensive legal analysis format
        legal_prompt = f"""As a legal AI assistant, analyze the following legal matter and provide comprehensive legal analysis including relevant laws, potential issues, risks, and recommendations.

Legal Matter:
{legal_question}

Legal Analysis:"""

        messages = [
            {"role": "user", "content": legal_prompt},
            {"role": "assistant", "content": legal_response}
        ]

        # Apply Qwen2.5 chat template
        formatted_text = tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=False,
        )

        texts.append(formatted_text)

    return {"text": texts}
```

```python
formatted_dataset = dataset.map(
    format_legal_analysis,
    batched=True,
    remove_columns=dataset.column_names,
    desc="Formatting legal advice data"
)
```

```python
from transformers import TrainingArguments
training_args = TrainingArguments(
    per_device_train_batch_size=1,  # Small batch for 14B model
    gradient_accumulation_steps=8,  # Effective batch size: 8
    warmup_steps=5,
    max_steps=80,  # Fewer steps for large model efficiency
    learning_rate=5e-5,  # Very low LR for large model stability
    fp16=not torch.cuda.is_bf16_supported(),
    bf16=torch.cuda.is_bf16_supported(),
    logging_steps=5,
    optim="adamw_8bit",
    weight_decay=0.01,
    lr_scheduler_type="cosine",
    seed=3407,
    output_dir="qwen25_legal_counsel",
    save_strategy="no",
    dataloader_num_workers=0,
)
```

```python
from trl import SFTTrainer
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=formatted_dataset,
    args=training_args,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    packing=False,
)
```

```python
trainer.train()
```

```python
def legal_consultation(legal_question, max_tokens=800):
    """Provide legal consultation based on r/legaladvice style training"""

    legal_prompt = f"""As a legal AI assistant, analyze the following legal matter and provide comprehensive legal analysis including relevant laws, potential issues, risks, and recommendations.

Legal Matter:
{legal_question}

Legal Analysis:"""

    messages = [
        {"role": "user", "content": legal_prompt}
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
    )

    inputs = tokenizer([text], return_tensors="pt").to("cuda")

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_tokens,
            temperature=0.3,  # Conservative for legal accuracy
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id,
        )

    generated_tokens = outputs[0][len(inputs['input_ids'][0]):]
    response = tokenizer.decode(generated_tokens, skip_special_tokens=True)

    return response.strip()
```

```python
employment_cases = [
    """
    My employer is making me work 60+ hours per week without overtime pay. They claim I'm "salary exempt" but I mostly do administrative tasks like data entry and filing. I make $35,000/year. They also dock my pay when I'm late even though they say I'm salaried. Is this legal? I'm in California.
    """,

    """
    I was fired yesterday after reporting safety violations to OSHA. My boss said it was for "performance issues" but my reviews have always been good. The safety issues I reported involved exposed electrical wires and blocked emergency exits. I have emails showing I reported this internally first. Do I have a case for wrongful termination?
    """,

    """
    My company wants me to sign a non-compete agreement that would prevent me from working anywhere in my industry for 2 years within 100 miles. I live in a small city where this industry is the main employer. They're saying if I don't sign, they'll find a reason to fire me. Can they do this? I'm in Texas.
    """
]

print("👔 EMPLOYMENT LAW CONSULTATION:")
for i, case in enumerate(employment_cases, 1):
    print(f"\n📋 Employment Issue {i}:")
    print(f"Question: {case[:100]}...")
    analysis = legal_consultation(case, 700)
    print(f"⚖️ Legal Analysis: {analysis}")
    print("-" * 60)
```
